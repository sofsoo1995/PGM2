 ## 1 Conditional independence and factorizations
 
 ### 1 let prove that $X \perp Y | Z$ iff $p(x|y,z)=p(x|z)$
 suppose $p(x |y,z)=p(x|z)$
we have 
\begin{align}
p(x,y|z) &= \frac{p(x,y,z)}{p(z)}\\
         &= \frac{p(x|y,z)p(y,z)}{p(z)}\\
         &= p(x|z)p(y|z)
\end{align}
So 
$$
\boxed{X \perp Y | Z}
$$
 Suppose now $X \perp Y |Z$
 we have 
 
\begin{align}
p(x|y,z) &= \frac{p(x,y,z)}{p(y,z)}\\
         &= \frac{p(x,y|z)p(z)}{p(y,z)}\\
         &= \frac{p(x|z)p(y|z)}{p(y|z)}\\
         &= p(x|z)
\end{align}
So,
$$
\boxed{p(x|y,z) = p(x|z)}
$$

 ### 3a) Let show that if $X$ and $Y$ are independant and $X$ and $Y$ are independant given $Z$ then $X$ independant to $Z$ or $Y$ independant to $Z$ , $Z$ is a binary random variable
 
 Let suppose that $X$ and $Y$ are independant and $X$ and $Y$ are independante given $Z$.
 
We have :
\begin{align}
p(x,y) &= p(x)p(y) \\
&= p(x,y|z=0)p(z=0)+p(x,y|z=1)p(z=1)
\end{align}
 let $\theta=p(z=1)$
 So
 \begin{align}
 p(x)p(y)&= p(x,y|z=0)(1-\theta)+p(x,y|z=1)\theta \\
         &=  p(x|z=0)p(y|z=0)(1-\theta)+p(x|z=1)p(y|z=1)\theta \\
         &= \frac{p(z=0|x)p(x)p(z=0|y)p(y)}{1-\theta}+\frac{p(z=1|x)p(x)p(z=1|y)p(y)}{\theta}
\end{align}
let $u = p(z=1|x)$ and $v=p(z=1|y)$
so
 \begin{align}
 1 &= \frac{(1-u)(1-v)}{1-\theta} + \frac{uv}{\theta} \\
   &= \frac{\theta(1-u)(1-v)+(1-\theta)uv}{\theta(1-\theta)}\\
   &= \frac{\theta(1+uv-u-v)+uv -\theta uv }{\theta(1-\theta)}\\
   &= \frac{uv -\theta u - \theta v + \theta}{\theta(1-\theta)}
 \end{align}
 Now let suppose that we don't have one independance for example $p(z=1|x) \neq p(z=1)$ or $u \neq \theta$
 So let show that $v=\theta$
 We have
$$
 \theta(1-\theta) = v(u-\theta) + \theta(1-u)
$$
so
$$
v(u-\theta) = \theta(1-\theta-1+u)
$$
But as we supposed : $u-\theta \neq 0$ so we can divide by $u-\theta$
We infer that 
$$
v = \theta
$$

## 2 Distributions factorizing in a graph

### 1 Show that $\mathcal{L}(G)=\mathcal{L}(G')$ for \{i,j\} a covered edge
let $p \in \mathcal{L}(G)$

We have
$$
p(x) = \prod_{k=1}^n p(x_k|x_{\pi_k}) = \prod_{k=1,k \neq i, k \neq j}^n p(x_k|x_{\pi_k}) p(x_i|x_{\pi_i})p(x_j|x_{\pi_j})
$$

but we suppose ${i \to j}$ is a covered edge. So $\pi_j=\pi_i \cup \{i\}$
So
$$
p(x_i|x_{\pi_i})p(x_j|x_{\pi_j}) = p(x_i|x_{\pi_i})p(x_j|x_{\pi_i},x_i)=\frac{p(x_j,x_{\pi_i},x_i)p(x_i,x_{\pi_i})}{p(x_{\pi_i},x_i)p(x_{\pi_i})} = p(x_i,x_j|x_{\pi_i})
$$
But we also have
$$
p(x_j|x_{\pi_i})p(x_i|x_{\pi_i},x_j) =p(x_i,x_j|x_{\pi_i})
$$
But this is the factorisation of $\mathcal{L}(G')$
So 
$$
\boxed{\in \mathcal{L}(G) \subset \mathcal{L}(G')}
$$
We have the other inclusion by the same way as the first inclusion(it is just the same calculus).

So 
$$
\boxed{\in \mathcal{L}(G) = \mathcal{L}(G')}
$$
