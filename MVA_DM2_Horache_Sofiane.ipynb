## 3 Entropy and Mutual Information

### 1a) prove that $H(X) \geq 0$
We have
$$
H(X) = \sum_{x \in mathcal{X}} p_X(x)\log(p_X(x))
$$
and 
$$
\forall x \in \mathcal{X} -p_X(x)\log(p_X(x)) \geq 0
$$
the fuction $f:x \to x\log(x)\geq 0 \forall x \in [0,1]$
$$
f(x)=0 \iff x \in \{0,1\}
$$
so 
$$
H(X)=0 \iff \exists! x_0 p_X(x_0)=1, \forall x \neq x_0 p_X(x)=0 
$$
So 
X is constant with probability 1.

### 1b) relations between $D(p_X||q)$ and $H(X)$ if $|\mathcal{X}|=k$
$q$ is a uniform distribution so :
$$
\forall x, q(x) = \frac{1}{k}
$$
Hence
\begin{align}
D(p_X||q) &= \sum_{x \in \mathcal{X}}p_X(x)\log(\frac{p_X(x)}{q(x)})\\
          &= \sum_{x \in \mathcal{X}}p_X(x)\log(kp_X(x))\\
          &=\sum_{x \in \mathcal{X}}p_X(x)\log(k) + \sum_{x \in \mathcal{X}}p_X(x)\log(p_X(x))\\
          &= \log(k) - H(X)
\end{align}

### 1c)upper bound of $H(X)$

We have
$$
D(p_X||q) \geq 0
$$
So 
$$
\log(k)-H(X) \geq 0
$$
Hence
$$
\boxed{H(X) \leq \log(k)}
$$

### 2a) prove that $I(X_1,X_2) \geq 0$

We have this because
$$
I(X_1,X_2) = D(p_{1,2}||p_1 p_2)\geq 0
$$
### 2b) $I(X_1,X_2)$ in function of $H(X_1)$ and $H(X_2)$
We write
\begin{align}
I(X_1,X_2) &= \sum_{(x_1,x_2) \in \mathcal{X}_1 \times \mathcal{X}_2} p_{1,2}(x_1,x_2) \log(\frac{p_{1,2}(x_1,x_2)}{p_1(x_1)p_2(x_2)})\\
           &= \sum_{(x_1,x_2) \in \mathcal{X}_1 \times \mathcal{X}_2} p_{1,2}(x_1,x_2)\log(p_{1,2}(x_1,x_2)) - \sum_{(x_1,x_2) \in \mathcal{X}_1 \times \mathcal{X}_2} p_{1,2}(x_1,x_2)\log(p_1(x_1)p_2(x_2))\\
           &= H(X_1,X_2) -\sum_{(x_1 \in \mathcal{X}_1} \sum_{x_2 \in \mathcal{X}_2} p_{1,2}(x_1,x_2)\log(p_2(x_2)) -\sum_{(x_1 \in \mathcal{X}_1}\sum_{x_2 \in \mathcal{X}_2} p_{1,2}(x_1,x_2)\log(p_1(x_1))\\
           &= H(X_1,X_2) - H(X_1)-H(X_2)
\end{align}

### 2c) joint distribution of maximal entropy
let call $k_1=|\mathcal{X}_1|$ and $k_2=|\mathcal{X}_2|$
We have according to question 1c
$$
H(X_1,X_2) \leq \log(k_1k_2) = \log(k_1)+\log(k_2)
$$
the equality hold when $p_{1,2}$ is a uniform distribution
But if $p_1$ and $p_2$ follows a uniform distribution, 
$$
 \log(k_1)+\log(k_2)=H(X_1)+H(X_2) \leq H(X_1,X_2) \leq \log(k_1)+\log(k_2)
$$
Hence $p_{1,2}$ follows a uniform distribution and
$$
\boxed{p_{1,2}(x_1,x_2)=p_1(x_1)p_2(x_2)}

### 1) K-means algorithm

```python
def kmeans(X, K, num_class=4):
    """this function cluster unlabeled data using the kmeans algorithm 
    input X: the point to cluster, num_class : the numbr of class, K : number of iteration
    
    output : a label for each point and the centroid"""
    # random initialisation
    
    centroid = np.random.permutation(X)[:num_class]
    label=np.zeros(X.shape[0])
    for i in range(X.shape[0]):
        score = np.array(list(map(lambda c:np.linalg.norm(X[i]-c),centroid)))
        label[i] = np.argsort(score)[0]
    for _ in range(K):
        # update centroid
        for i in range(num_class):
            centroid[i] = np.mean(X[label==i],axis=0)
        #update label
        for i in range(X.shape[0]):
            score = np.array(list(map(lambda c:np.linalg.norm(X[i]-c),centroid)))
            label[i] = np.argsort(score)[0]
            
    #print the "distortion" measure ie the variance of the distance
    distortion = 0
    for i in range(num_class):
        Xc = X[label==i]-centroid[i]
        
        distortion += np.trace(np.dot(Xc,Xc.T))/np.sum(label==i)
    print("distortion : %0.3f" % distortion)
    return label, centroid

def plot_kmeans():
    """this function just plot the result of kmeans"""
    label,centroid = kmeans(X_train,30)

    figure = plt.figure(figsize=(10,10))
    for i in range(4):
        plt.scatter(X_train[label==i][:,0], X_train[label==i][:,1])
    plt.scatter(centroid[:,0], centroid[:,1],c='black',marker='+',s=200)
    plt.legend(['class 1', 'class 2', 'class 3', 'class 4'])
    plt.show()
```

```python
def normal(x,mu,sigma):
    d = sigma.shape[0]
    xc = x-mu;
    xct = np.linalg.inv(sigma).dot(xc)
    return 1.0/np.sqrt((2*np.pi)**d * np.linalg.det(sigma)) * np.exp(-0.5*xc.dot(xct))



def GMM(X, K, num_class=4,isotropic=True):
    """this function apply the EM algorithm to estimate the parameters of the gmm"""
    
    # initialization with kmeans
    label,mu=kmeans(X, 10, num_class=num_class)
    pi = [ np.sum(label==i) for i in range(num_class)]
    sigma = [np.eye(X.shape[1]) for i in range(num_class)] # initialisation with identity matrix
    normals = np.zeros((X.shape[0],num_class))
    Q = np.zeros((X.shape[0],num_class))
    for i in range(num_class):
        normals[:,i] = np.array([normal(X[j],mu[i],sigma[i]) for j in range(X_train.shape[0])])
    for i in range(X.shape[0]):
        Q[i,:]=normals[i,:]*np.array(pi)/np.dot(np.array(pi),normals[i,:])
    
    for _ in range(K):
        # Maximization
        for i in range(num_class):
            Q2 = np.zeros(X.shape)
            Q2[:,0]=Q[:,i]
            Q2[:,1]=Q[:,i]
            mu[i] = np.sum(X*Q2,axis=0)/np.sum(Q[:,i])
            pi[i] = np.sum(Q[:,i])/ np.sum(np.sum(Q,axis=0))
            if(isotropic):
                sigma[i] = np.dot(np.diag(np.cov(X-mu[i])),Q[:,i])/np.sum(Q[:,i]) * np.eye(X.shape[1])
            else:
                Xc=X-mu[i]
                sigma[i] = Xc.T.dot(Q2*Xc)/np.sum(Q[:,i])
    
    # Expectation
    normals = np.zeros((X.shape[0],num_class))
    Q = np.zeros((X.shape[0],num_class))
    for i in range(num_class):
        normals[:,i] = np.array([normal(X[j],mu[i],sigma[i]) for j in range(X.shape[0])])
    #for i in range(X.shape[0]):
     #   Q[i,:]=normals[i,:]*np.array(pi)/np.dot(np.array(pi),normals[i,:]) 
    Q = normals * np.repeat(np.array([pi]),X.shape[0],axis=0)/np.repeat(
        np.array([normals.dot(np.array(pi))]).T, normals.shape[1],axis=1)
    #compute probability for each Xi to belong to a class. it multiply each proba of Xi with his pi and then normalize
    return pi,mu,sigma,Q


def compute_log_likelihood(X,pi,mu,sigma,Q):
    normals = np.zeros((X.shape[0],Q.shape[1]))
    for i in range(num_class):
        normals[:,i] = np.array([np.log(normal(X[j],mu[i],sigma[i])) for j in range(X.shape[0])])
    
    res = np.trace(normals.T.dot(Q)) + np.sum(Q.dot(np.log(pi))) 
    return res

def cluster(X,pi,mu,sigma):
    """this function clusters softly a given point x according to the parameter"""
    normals = np.zeros((X.shape[0],num_class))
    Q = np.zeros((X.shape[0],num_class))
    for i in range(num_class):
        normals[:,i] = np.array([normal(X[j],mu[i],sigma[i]) for j in range(X.shape[0])])
    Q = normals * np.repeat(np.array([pi]),X.shape[0],axis=0)/np.repeat(
        np.array([normals.dot(np.array(pi))]).T, normals.shape[1],axis=1)
    return Q
    
def draw_ellipse(mui,sigmai):
    val, vec = np.linalg.eigh(sigmai)
    order = val.argsort()[::-1]
    val, vec = val[order],vec[order]
    
    width, height = 2 * 1.96 * np.sqrt(val)
    theta = np.degrees(np.arctan2(*vec[:,1][::-1]))
    e = Ellipse(xy=mui,width=width, height=height, angle=theta)
    return e 
def plot_gmm(X,pi,mu,sigma,Q):
    """this function compute the GMM and then draw points ellipse and print the loglikelihood on the title"""
    label = np.argmax(Q,axis=1)
    #figure = plt.figure(figsize=(10,10))
    fig, subplot = plt.subplots(subplot_kw={'aspect': 'equal'},figsize=(10,12))
    for i in range(4):
        e = draw_ellipse(mu[i], sigma[i])
        e.set_clip_box(subplot.bbox)
        subplot.scatter(X[label==i][:,0], X[label==i][:,1])
        e.set_alpha(0.5)
        subplot.add_artist(e)
    subplot.scatter(mu[:,0], mu[:,1],c='black',marker='+',s=200)
    plt.legend(['class 1', 'class 2', 'class 3', 'class 4'])
    plt.title('log likelihood of gmm : %0.3f'%compute_log_likelihood(X,pi,mu,sigma,Q))
    plt.show()
def plot_latent(X,pi,mu,sigma,res=20):
    """this function will represent the latent variable for all point with the GMM res is the resolution"""
    Xs = np.linspace(-10,8,res)
    Ys = np.linspace(-10,12,res).T
    Xs,Ys = np.meshgrid(Xs,Ys)
    pairs = np.c_[Xs.ravel(), Ys.ravel()]
    Q = cluster(pairs,pi,mu,sigma)
    label = np.argmax(Q,axis=1)
    label = label.reshape(Xs.shape)
    figure = plt.figure(figsize=(10,10))
    plt.pcolormesh(Xs,Ys,label,cmap='plasma',alpha=0.5,shading='gouraud')
    plt.legend(['class 1', 'class 2', 'class 3', 'class 4'])
    plt.scatter(X[:,0],X[:,1])
    plt.title('areas where p(z|x) is higher')
    plt.show()
```
